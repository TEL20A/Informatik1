unterschied double vs. float

```c++
// Größe der unterschiedlichen Datentypen
std::string langer_string = "12345";
std::vector<std::string> board = {
    "123",
    "456",
    "789"
};

std::cout << "Größe der Datentypen in Byte:" << std::endl;

std::cout << "double: " << sizeof(double) << std::endl;
std::cout << "float: " << sizeof(float) << std::endl;

std::cout << "int: " << sizeof(int)     << std::endl;
std::cout << "short: " << sizeof(short) << std::endl;
std::cout << "long: " << sizeof(long)   << std::endl;
std::cout << "char: " << sizeof(char)   << std::endl;
std::cout << "String: " << sizeof(std::string)   << std::endl;
std::cout << "Langer String: " << sizeof(langer_string)   << std::endl;
std::cout << "groesse board: " << sizeof(board)   << std::endl;

// short hat 65536 verschiedene Werte 0 - 65536
// bei einem signed short teilen in den negative und posistiven Bereich
short s_short = 32765;
for(int i=0; i < 10; i++) {
  std::cout << s_short << std::endl;
  s_short++;
}
std::cout << "Unsigned short:\n";
unsigned short us_short = 65530;
for(int i=0; i < 10; i++) {
  std::cout << us_short << std::endl;
  us_short++;
}


std::cout << "char:\n";
char s_char = 30;
for(int i=0; i < 10; i++) {
    int temp = s_char;
    std::cout << temp << std::endl;
    s_char++;
}
```

## Typedef
Mit typedef hat man die Möglichkeit einen neuen Namen für einen Datentypen zu vergeben.

Z.B. erstelle ich den neuen Datentyp `clock_t` der aus dem Typ `long int` besteht.

```c++
typedef unsigned long clock_t;
clock_t time;
time = 250;
std::cout << "time: " << time << std::endl;

std::cout << "clock_t: " << sizeof(clock_t)   << std::endl;
std::cout << "long: " << sizeof(long)   << std::endl;
```

```c++
enum sex_type {
    mann, //0
    m=0, 
    frau, //1
    f=1, 
    divers //2
};

// Definition eines neuen Datentyps person
struct person {
    std::string name;
    short alter;
    enum sex_type geschlecht;
    
    //Methoden
    void set(std::string _name, short _alter, enum sex_type _geschlecht) {
        this->name = _name;
        this->alter= _alter;
        this->geschlecht = _geschlecht;
    }
    // Der this operator ermöglicht es auf im struct definierte Variablen zuzugreifen
    // WICHTIG: Der Pfeil (->) bei this-> zum Variablenzugriff
    void print() {
        std::cout << "name: " << this->name << std::endl;
        std::cout << "alter: " << this->alter << std::endl;
        std::cout << "geschlecht: " << this->geschlecht << std::endl;
    }
};

struct person PersonA;
PersonA.name = "studentA";
PersonA.alter = 22;
PersonA.geschlecht = divers;

std::cout << "PersonA.name: " << PersonA.name << std::endl;
std::cout << "PersonA.alter: " << PersonA.alter << std::endl;
std::cout << "PersonA.geschlecht: " << PersonA.geschlecht << std::endl;

std::cout << "Benutzung der Methode print von struct person " << std::endl;
PersonA.print();

struct person PersonB;
PersonB.name = "studentB";
PersonB.alter = 25;
PersonB.geschlecht = mann;

PersonB.print();
struct person PersonC;
PersonC.set("PersonC", 26, frau);
PersonC.print();


//Typedef mit struct zum Definieren eines neuen eigenen Datentypen
typedef struct person person;
person PersonD;
PersonD.set("PersonD", 26, frau);
PersonD.print();
```
